## Addition function

```haskell
add :: Integer -> Integer -> Integer -> Integer
add x y z = x + y + z                         
```

```haskell
add 1 2 3
```

## Factorial function

```haskell
fact :: Int -> Int
fact 0 = 1
fact n = n * fact (n - 1)
```

```haskell
fact 5
```

## Solving quadratic equation with real roots

```haskell
roots :: (Float, Float, Float ) -> (Float, Float)
roots (a,b,c) = (x1, x2) where
  x1 = e + sqrt d / (2 * a)
  x2 = e - sqrt d / (2 * a )
  d = b * b - 4 * a * c 
  e = - b / ( 2 * a)
```

```haskell
roots (1,2,1)
```

## Using the discriminant to find the number of roots and solving for the roots

```haskell
disc :: Float->Float->Float->IO()

disc a b c = do
  print ("a,b,c values=",a,b,c)
  let d =b^2 -4*a*c
  let e = - b / (2 * a)
  let root1 = e + sqrt d / (2 * a)
  let root2 = e - sqrt d / (2 * a)
  if d < 0
    then do
      print "this quadratic has no real roots"
      print (root1, root2)
  else if d == 0
    then do
      print "this quadratic has one repeating root"
      print root1
  else 
    do
      print "this quadratic has two real roots"
      print (root1, root2)
  putStrLn " the discriminant of the quadratic is:"
  print d
```

```haskell
disc 6 15 2
```

## Choose function $C_p^n$  
This is the number of ways to choose p items from n items without regard to order.

```haskell
choose :: Int -> Int -> Int
choose n p = (fact n) `div` ((fact p) * (fact (n - p)))
```

```haskell
choose 5 2
```
