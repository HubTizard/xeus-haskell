# Q4
There are 13 marbles in a jar. The jar can hold 20 marbles. How many more marbles can fit in the jar?

```haskell
type Marbles = Int
```

```haskell
outside = 13 :: Marbles
```

```haskell
jar = 20 :: Marbles
```

```haskell
more = jar - outside
```

```haskell
more
```

# Q5
Write the same __number__ in both boxes to make the multiplication correct: [ ] * [ ] = 25

## Solution 1, recall square roots (not in national curriculum for Year 2)

## Solution 2, keep squaring integers greater than or equal to 1 until you reach 25. Then count how many numbers you have
why do I use the length function? why do I need to add 1?

```haskell
(takeWhile (\x -> x * x /= 25) [1..])
```

```haskell
length (takeWhile (\x -> x * x /= 25) [1..]) + 1
```
